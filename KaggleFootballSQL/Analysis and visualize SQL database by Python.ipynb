## As the first target is to identify all the match of Manchester United in season 2011/2012 which is the last season of Sir Alex Ferguson, we have 

```python
MU_home_result_2012 = pd.read_sql(
"""SELECT m.date, t.team_long_name AS opponent,
    CASE WHEN m.home_team_goal > m.away_team_goal THEN "MU win"
         WHEN m.home_team_goal < m.away_team_goal THEN "MU lose"
         ELSE 'TIE' END AS Result
FROM Match as m LEFT JOIN Team AS t
ON m.away_team_api_id = t.team_api_id
WHERE m.home_team_api_id = 10260
AND m.season = '2011/2012'""", connection)

MU_home_result_2012
```

```python
MU_home_result_2012['Result'].value_counts().plot.bar(color = ['blue','yellow','red'], alpha = 0.2)
plt.ylabel('Number')
plt.show()
```

What about the result of derby between Manchester United and Manchester City ? We knows that the Id of Manchester United is 10260 and Manchester City is 8456. Let's do a analysis to see which team performce better compare to the others

```python
MU_vs_Mancity = pd.read_sql("""SELECT date, season, 
CASE WHEN home_team_api_id = 10260 THEN 'Manchester United'
    ELSE 'Manchester city' END AS home,
CASE WHEN away_team_api_id = 10260 THEN 'Manchester United' 
    ELSE 'Manchester city' END AS away,

CASE WHEN home_team_goal > away_team_goal AND home_team_api_id = 10260 THEN 'MU win'
     WHEN home_team_goal > away_team_goal AND home_team_api_id = 8456 THEN 'Mancity win'
     WHEN home_team_goal < away_team_goal AND home_team_api_id = 10260 THEN 'MU win'
     WHEN home_team_goal < away_team_goal AND home_team_api_id = 8456 THEN 'Mancity win'
     ELSE 'Tie' END AS Result
FROM Match 
WHERE(home_team_api_id = 10260 OR away_team_api_id = 10260) 
AND (home_team_api_id = 8456 OR away_team_api_id = 8456)
ORDER BY home""", connection)

MU_vs_Mancity
```

```python
MU_vs_Mancity['Result'].value_counts().plot.bar(color = ['blue','red','green'], alpha = 0.2)
plt.ylabel('Number')
plt.show()
```

## How about comparing the total goals of each league in season 2015/2016 ? 

Firstly, we have to create a basic query to extract the data in season 2015/2016 in 5 most popular league

```python
league_2016 = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                AND season = '2015/2016'
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
league_2016
```

```python
plt.plot
plt.bar(league_2016['country_name'],league_2016['total_goals'], color = ['green','blue','yellow','violet','purple'], alpha = 0.3)
plt.show()
```

```python
league_season = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
league_season.head(10)
```

```python
goals = league_season.groupby(['country_name','season']).sum()[['total_goals']] # all the goal of each country 
```

```python
plt.plot(goals.loc['England',:].values, label = 'England')
plt.plot(goals.loc['France',:].values, label = 'France')
plt.plot(goals.loc['Germany',:].values, label = 'Germany')
plt.plot(goals.loc['Italy',:].values, label = 'Italy')
plt.plot(goals.loc['Spain',:].values, label = 'Spain')
plt.xticks([0,1,2,3,4,5,6,7],['2008/2009','2009/2010','2010/2011','2011/2012','2012/2013','2013/2014','2014/2015','2015/2016'])
plt.legend()
plt.show()
```
